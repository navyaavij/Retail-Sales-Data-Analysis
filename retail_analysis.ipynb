```python
import pandas as pd
import numpy as np

df = pd.read_csv("Sample-Superstore.csv")

df.head()
```

```python
df.fillna("Unknown", inplace = True)
```

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]

print(f"Removed {before - after} duplicate rows")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.dtypes
```

```python
df["Shipping_Duration"] = (df["Ship Date"] - df["Order Date"]).dt.days
df[["Order Date", "Ship Date", "Shipping_Duration"]].head()
```

```python
df["Order_Month_Name"] = df["Order Date"].dt.month_name()
```

```python
df["Profit_Margin"] = df["Profit"] / df["Sales"]
df[["Sales", "Profit", "Profit_Margin"]].head()
```

NUMPY ANALYSIS


```python
sales_array = df["Sales"].values
np.mean(sales_array)
```

```python
np.median(sales_array)
```

```python
np.std(sales_array)
```

### Outlier Analysis

q1 = np.percentile(sales_array, 25)
q3 = np.percentile(sales_array, 75)

iqr = q3 - q1
upper_limit = q3 + 1.5 * iqr

upper_limit

```python
outliers = sales_array[sales_array > upper_limit]
len(outliers)
```

Using the IQR method, sales values above 573.64 were identified as outliers.
A total of 1,164 high-value orders were found and retained, as they represent legitimate business transactions.

INSIGHTS

- November recorded the highest total sales, indicating strong seasonal demand.
  
- A negative correlation was observed between Discount and Profit, suggesting that higher discounts reduce overall profitability.
